```python
fig = plt.figure(figsize=(12, 8))
cuantitativas.plot (kind = 'box', vert = False)
plt.title("Valores Atípicos del DataFrame")
plt.show()
```

```python
fig = plt.figure(figsize=(15, 8))
cuantitativas["SUBTOTAL_PARTIDA"].plot (kind = 'box', vert = False)
plt.title("Valores Atípicos del DataFrame")
plt.show()
```

```python
y= cuantitativas
Limite_Superior = y.mean() + 3 * y.std()
Limite_Inferior = y.mean() - 3 * y.std()
print("Limite Superior Permitido",Limite_Superior)
print("Limite Inferior Permitido",Limite_Inferior)
```

```python
data2= cuantitativas[(y<=Limite_Superior) & (y>=Limite_Inferior)]
data2
```

```python
valores_nulos = data2.isnull().sum()
valores_nulos
```

```python
#Remplazo valores atipicos (nulos) del dataframe con "mean"
#Realizamos una copia del data
data_clean= data2.copy()
data_clean = data_clean.fillna(round(data2.mean(),1))
data_clean
```

```python
valores_nulos=data_clean.isnull().sum()
valores_nulos
```

```python
y=cuantitativas
percentiles25 = y.quantile(0.25) #Q1
percentiles75 = y.quantile(0.75) #Q3
iqr = percentiles75 - percentiles25

Limite_Superior_iqr = percentiles75 + 1.5 * iqr
Limite_Inferior_iqr = percentiles25 - 1.5 * iqr
print("Limite Superior Permitido",Limite_Superior_iqr)
print("Limite Inferior Permitido",Limite_Inferior_iqr)
```

```python
data3_iqr = cuantitativas[(y<=Limite_Superior_iqr) & (y>=Limite_Inferior_iqr)]
data3_iqr
```

```python
valores_nulos = data3_iqr.isnull().sum()
valores_nulos
```

```python
data4_iqr= data3_iqr.copy()
data4_iqr = data4_iqr.fillna(round(data3_iqr.mean(),1))
data4_iqr
```

```python
Data_limpios = pd.concat([cualitativas, data4_iqr], axis=1)
Data_limpios
```

```python
valores_nulos = Data_limpios.isnull().sum()
valores_nulos
```
